```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('netflix_titles.csv')
```

```python
# Cleaning
df.drop_duplicates(inplace=True)
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna('Not Rated', inplace=True)
df['date_added'].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['type'].value_counts()
```

```python
df['type'].value_counts().plot(kind='bar', color=['red','gray'])
plt.title('Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar', color='red')
plt.title('Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot(kind='line', color='red')
plt.title('Year wise Content')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
df['listed_in'].value_counts().head(10).plot(kind='bar', color='green')
plt.title('Top 10 Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

```python
df['rating'].value_counts().plot(kind='bar', color='orange')
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['red','gray'])
plt.title('Movies vs TV Shows')
plt.show()
```

```python
sns.countplot(x='type', data=df, palette='Set1')
plt.title('Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(y='rating', data=df, palette='Set2', order=df['rating'].value_counts().index)
plt.title('Rating Distribution')
plt.xlabel('Count')
plt.ylabel('Rating')
plt.show()
```

```python
sns.barplot(x=df['country'].value_counts().head(10).index,
            y=df['country'].value_counts().head(10).values,
            palette='Set1')
plt.title('Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.histplot(df['release_year'], color='red', bins=20)
plt.title('Year wise Content')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
sns.barplot(x=df['listed_in'].value_counts().head(10).values,
            y=df['listed_in'].value_counts().head(10).index,
            palette='Set2')
plt.title('Top 10 Genres')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```
